```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
r_values = np.linspace(2.5,4,10000)

y_values = []

for r in r_values:
    temp_list = []
    x_values = np.linspace(0,1,400)
    for x_0 in x_values:
        x = x_0
        for k in range(400):
            x = r * x * (1-x)
        temp_list.append(x)
    y_values.append(temp_list)

big_y = []
for k in y_values:
    big_y.append(k)
big_r = []
for r in r_values:
    for k in range(len(y_values[0])):
        big_r.append(r) 
```

```python
plt.scatter(big_r,big_y, s=0.0001, alpha = 0.25)

plt.savefig("logistic_map_graph.png")
```
